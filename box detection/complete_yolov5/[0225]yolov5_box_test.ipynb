```python
def show(img):
    #사이즈
    plt.figure(figsize = (50,30))
    #xticks/yticks - 눈금표
    plt.xticks([])
    plt.yticks([])
    #코랩에서 안돌아감 주의
    plt.imshow(img, cmap= 'gray')
    plt.show()
```

```python
import cv2
import matplotlib.pylab as plt
img = cv2.imread('/content/drive/MyDrive/20220126_172410.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# img = cv2.resize(img, dsize = (832, 1024))
show(img)
img.shape
```

```python
cord=[]
f = open('/content/drive/MyDrive/save_box/test.txt', 'r')
lines = f.readlines()
for line in lines:
    k = line.split('\n')[0]
    k = k.replace('(','')
    k = k.replace(')','')
    k = k.replace(',','')
    cord.append(k)
cord
```

```python
x = int(cord[0].split(' ')[0])
y = int(cord[0].split(' ')[1])
w = int(cord[0].split(' ')[2])
h = int(cord[0].split(' ')[3])
```

```python
import math
```

```python
img_ =img.copy()
 
for i in cord:
   
    x = int(i.split(' ')[0])
    y = int(i.split(' ')[1])
    w = int(i.split(' ')[2])
    h = int(i.split(' ')[3])

    ctx= math.trunc((x+x+w)/2)
    cty = math.trunc((y+y+h)/2)
    

    # show(img[y:y+h, x:x+w])

    img_ = cv2.circle(img_, (ctx,cty), 20, (0,0,255), -1)
    # print(i)
show(img_)

    # # # show(img[y:y+h, x:x+h])
```
