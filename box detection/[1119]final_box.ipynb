```python
import torch
import cv2
import os
import matplotlib.pyplot as plt
```

```python
def show(img):
    #사이즈
    plt.figure(figsize = (10,8))
    #xticks/yticks - 눈금표
    plt.xticks([])
    plt.yticks([])
    #코랩에서 안돌아감 주의
    plt.imshow(img, cmap= 'gray')
    plt.show()
```

```python
test = cv2.imread('/Users/kimmonica/Desktop/frame_2.png', cv2.IMREAD_COLOR)
test = cv2.cvtColor(test, cv2.COLOR_BGR2RGB)
print(test.shape)
show(test)
```

```python
img_path = '/Users/kimmonica/Desktop/kor_rus/ghgh/'
img_list = os.listdir(img_path)

list_img = []

for i in img_list:
    path = img_path+i
    try:
        img = cv2.imread(path, cv2.IMREAD_COLOR)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        list_img.append(img)
    except:
        pass
    
print(list_img[0].shape)
show(list_img[0])
```

```python
src_img = []

for num, i in enumerate(list_img):
    try:
        src = i.copy()
        src = src[0:950, 350:1400]
        img = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
        save_path = '/Users/kimmonica/Desktop/kor_rus/resized/'+ img_list[num]
        print(save_path)
        cv2.imwrite(save_path, img)
    except:
        pass
```
